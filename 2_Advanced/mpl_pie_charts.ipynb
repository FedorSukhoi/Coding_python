```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import ast

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.2f%%')
plt.title('Work from Home Status')
plt.ylabel('')
plt.show()
```

```python
df[['job_work_from_home', 'job_no_degree_mention', 'job_health_insurance']]
```

```python
fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_work_from_home': 'Work from home',
    'job_no_degree_mention': 'Degree Necessity',
    'job_health_insurance': 'Health Insurance'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(
        df[column].value_counts(), 
        startangle=90, 
        autopct='%1.1f%%', 
        labels=['False', 'True'], 
        textprops={'fontsize': 8})
    ax[i].set_title(title)

plt.show()
```

```python
df_us = df[df['job_location'] == 'United States']
```

```python
fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_work_from_home': 'Work from home',
    'job_no_degree_mention': 'Degree Necessity',
    'job_health_insurance': 'Health Insurance'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(
        df_us[column].value_counts(), 
        startangle=90, 
        autopct='%1.f%%',  
        textprops={'fontsize': 8})
    ax[i].set_title(title)

plt.show()
```
