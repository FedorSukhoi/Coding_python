**NumPy**

```python
import random

salary_list = [random.randint(50000, 100000) for _ in range(1000000)]
```

```python
import numpy as np
np.mean(salary_list)
```

```python
my_array = np.array(salary_list)
my_array.mean()
```

```python
my_array.std()
```

```python
ed_array = my_array[my_array > 80000]
ed_array.size
```

```python
ed_array.std()
```

```python
ed_array.mean()
```
