```python
job_skills = ['sql', 'r', 'excel']
```

```python
job_skills.append('python')
job_skills.append('java')
job_skills
```

```python
job_skills.remove('python')
job_skills
```

```python
job_skills[0]
```

```python
job_skills[3]
```

```python
job_skills.insert(1, 'tableu')
job_skills
```

**Slicing**

```python
job_skills = ['sql', 'r', 'excel']
job_skills[0:2]
```

```python
job_skills[:]
```

```python
job_skills[:2]
```

```python
job_skills[1:]
```

```python
job_skills[::2] #перечисление из списка, начиная с первого, дальше пропуская один
```

```python
job_skills[-1:]
```

**Unpacking**

```python
job_skills = ['sql', 'r', 'excel']
skill1, skill2, skill3 = job_skills
print(skill1)
print(skill2)
print(skill3)
```

```python
cool_skills, *useless_skills = job_skills
print(cool_skills)
print(useless_skills)
```

**Dictionaries**

```python
job_type_skills = {
    'database': ['postgresql', 'mysql'],
    'language': 'python',
    'library': 'pandas'
}
job_type_skills
```

```python
job_type_skills['database']
```

```python
job_type_skills.keys()
```

```python
job_type_skills.values()
```

```python
job_type_skills.update({'cloud': 'google cloud'})
job_type_skills
```

```python
job_type_skills['version_control'] = 'github' #другой способ делать апдейт, функционал как у команды выше
job_type_skills
```

**Sets**

```python
them_skills = ({'tableu', 'sql', 'python', 'advanced statistics'})
them_skills #часто выдает в алфавитном порядке, но на самом деле не служит для этого, есть команда sorted()
```

```python
sorted(them_skills)
```

```python
them_skills.add('looker')
them_skills
```

```python
them_skills.remove('sql')
them_skills.remove('tableu')
them_skills
```

```python
the_skills = ['python', 'sql', 'sql']
set(the_skills) #sets наиболее полезны в ситуациях, где нужны уникальные значения
```

```python
list(set(the_skills)) #и часто их используют просто как команду, а потом превращают обратно в нужный формат уже без дубликатов
```

**Tuples**

```python
my_skills = ('sql', 'r', 'tableu') #список вида Tuple не подлежит никаким изменениям, для этого он и нужен
my_skills
```

```python
my_skills[0:2]
```

```python
new_skill = ('excel', 'python')
all = new_skill + my_skills
all
```
