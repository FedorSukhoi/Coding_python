```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
dt_counts = df.job_posted_date.value_counts()
dt_counts.sort_index(inplace=True)
```

```python
plt.plot(dt_counts.index, dt_counts.values)
```

```python
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
monthly_counts = df.job_posted_month.value_counts()
monthly_counts = monthly_counts.sort_index()
monthly_counts
```

```python
plt.plot(monthly_counts.index, monthly_counts)
```

```python
count = df.job_title_short.value_counts()
```

```python
count = count.sort_values(ascending=True)
plt.barh(count.index, count)
plt.show()
```

```python
count2 = df.job_title_short.value_counts()
```

```python
plt.bar(count2.index, count2)
plt.title('Postings by Job Title')
plt.xlabel('Number of Postings')
plt.ylabel('Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
count.plot(kind='bar') #функция pandas, которую можно связывать с командами mpl
plt.title('Postings by Job Title')
plt.xlabel('Number of Postings')
plt.ylabel('Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
df = df.set_index('job_posted_date')
df.sort_index(inplace=True)
df.head()
```

```python
df.plot(y='salary_year_avg', kind='line')
```

```python
job_salary = df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
```

```python
job_salary.plot(kind='barh')
plt.title('Median Salary by Job Title')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
