```python
people_experience = {
    'Kelly': 7,
    'Milo': 3,
    'Kyle': 16,
    'Jack': 4
}
```

```python
for key, value in people_experience.items():
  if value > 5:
    print(key, 'has', value, 'years of experience')
  else:
    print(key, 'has', value, 'years of experience, but that is not enough')
```
