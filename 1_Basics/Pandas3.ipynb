```python
df['job_title_short'].value_counts()
```

```python
df.groupby('job_title_short')['salary_year_avg'].mean().sort_values(ascending=False) #средняя зп по каждой должности
```

```python
df.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False) #медианная зп по каждой должности
```

```python
df.groupby('job_title_short')['salary_year_avg'].min().sort_values(ascending=False) #самые маленькие зп
```

```python
df.groupby('job_title_short')['salary_year_avg'].max().sort_values(ascending=False) #самые большие зп
```

```python
df.groupby('job_title_short')['salary_year_avg'].idxmax().sort_values(ascending=False) #индексы самых больших зп
```

```python
df.groupby('job_title_short')['salary_year_avg'].idxmin().sort_values(ascending=False) #индексы самых маленьких зп
```

```python
df.groupby(['job_title_short', 'job_country'])['salary_year_avg'].median() #медианная зп по каждой стране
```

```python
df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].mean().sort_values(by='salary_year_avg', ascending=False) #средняя зп по каждой стране
```

```python
df.groupby('job_title_short')['salary_year_avg'].agg(['min', 'max', 'mean', 'median']).sort_values(by='median', ascending=True) #все зп
```
